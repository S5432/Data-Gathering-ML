# Finding Elements of Webpage Using Python.

```python
# how to get data and text of the html tag
# 1. FInding element  data by class.
# 2. FInding element  data by ID.

import requests
from bs4 import BeautifulSoup

web = requests.get("https://www.tutorialsfreak.com/")
web
```

```python
soup = BeautifulSoup(web.content,"html.parser")
```

```python
print(soup.prettify)
```

```python
soup.find_all("p")
```

# By Class Find the Element Tags Data.

```python
class_data = soup.find("div", class_="bg-white border-0 p-3 card-footer")
```

```python
print(class_data)
```

```python
class_data.find_all("a")
```

# By ID Find the element tags data.

```python
class_Id = soup.find("div", id="__next")
print(class_Id)
```

```python
class_Id.find_all("p")
```

# Extracting text from the html tags of the webpages .

```python
lines = soup.find_all("p")
lines
```

```python
for i in lines:
  print(i.text)
```

```python
# get the data/text of the perticular paragraph.
s = soup.find("p", class_ = "fs-16 fw-400 lh-24 label-color-1 card-text")
```

```python
print(s)
```

```python
lines = soup.find_all("p")
lines
```

```python
for i in lines:
  print(i.text)
```

# Shortcut way

# Extracting links of the webpages.

```python
m = soup.find_all("a")
m
```

```python
for i in soup.find_all("a"):
   print(i.get("href"))
```

# Extracting the image information from the webpages.

```python
img = soup.find_all("img")
img
```

```python
for i in img:
  print(i.get("src"))
  print(type(i.get("src")))
```
